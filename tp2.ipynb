Neste notebook, as três tranformações geométricas elementares são implementadas em OpenGL.
Comandos:
* Translação:
  * Setas do teclado: translação em X e em Y.
  * PageUP/PageDOWN: translação em Z.
* Escala:
  * s: reescala o objeto em 1/2 em todos os eixos.
  * s: reescala o objeto em 2 em todos os eixos.
* Rotação:
  * x: rotação em torno do eixo x.
  * X: rotação em torno do eixo x (sentido contrário).
  * y: rotação em torno do eixo y.
  * Y: rotação em torno do eixo y (sentido contrário).
  * z: rotação em torno do eixo z.
  * Z: rotação em torno do eixo z (sentido contrário).
* Mouse Left Click: alterna entre paralelepípedo e teapot

## Importa Bibliotecas

```python
import sys
from OpenGL.GLUT import *
from OpenGL.GL import *
from OpenGL.GLU import *
import numpy as np

angle_x = 0
angle_y = 0
angle_z = 0
draw_teapot = False
```

## Função Display

A função display irá desenhar um cubo na tela (ou um teapot). Ela também faz a tranformação de rotação em torno dos eixos x, y e z.

```python
def display():
    global angle_x, angle_y, angle_z, draw_teapot
    
    glMatrixMode(GL_MODELVIEW)
    
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)

    glColor3f(0.0, 0.0, 1.0)

    glPushMatrix()
    
    glRotatef(angle_x, 1, 0, 0)
    glRotatef(angle_y, 0, 1, 0)
    glRotatef(angle_z, 0, 0, 1)
    
    if draw_teapot:
        glutSolidTeapot(30.0)
    else:
        # Back face
        glBegin(GL_QUADS)
        glNormal3f(0.0, 0.0, -1.0)
        glVertex3f(  25, -15, -50 )
        glVertex3f(  25,  15, -50 )
        glVertex3f( -25,  15, -50 )
        glVertex3f( -25, -15, -50 )
        glEnd()

        # Front face
        glBegin(GL_QUADS)
        glNormal3f(0.0, 0.0, 1.0)
        glVertex3f(  25, -15, 50 )
        glVertex3f(  25,  15, 50 )
        glVertex3f( -25,  15, 50 )
        glVertex3f( -25, -15, 50 )
        glEnd()

        # Right face
        glBegin(GL_QUADS)
        glNormal3f(1.0, 0.0, 0.0)
        glVertex3f( 25, -15, -50 )
        glVertex3f( 25,  15, -50 )
        glVertex3f( 25,  15,  50 )
        glVertex3f( 25, -15,  50 )
        glEnd()

        # Left face
        glBegin(GL_QUADS)
        glNormal3f(-1.0, 0.0, 0.0)
        glVertex3f( -25, -15,  50 )
        glVertex3f( -25,  15,  50 )
        glVertex3f( -25,  15, -50 )
        glVertex3f( -25, -15, -50 )
        glEnd()

        # Top face
        glBegin(GL_QUADS)
        glNormal3f(0.0, 1.0, 0.0)
        glVertex3f(  25,  15,  50 )
        glVertex3f(  25,  15, -50 )
        glVertex3f( -25,  15, -50 )
        glVertex3f( -25,  15,  50 )
        glEnd()

        # Bottom face
        glBegin(GL_QUADS)
        glNormal3f(0.0, -1.0, 0.0)
        glVertex3f(  25, -15, -50 )
        glVertex3f(  25, -15,  50 )
        glVertex3f( -25, -15,  50 )
        glVertex3f( -25, -15, -50 )
        glEnd()

    glPopMatrix()

    glFlush()
```

## Funções para lidar com os inputs de teclado e mouse
As funções mouse_click, key_pressed_special e key_pressed lidam com os eventos de mouse/teclado.

```python
def mouse_click(button, state, x, y):
    global draw_teapot

    if state == GLUT_UP:
        return
    if button == GLUT_LEFT_BUTTON:
        draw_teapot = ~draw_teapot

def key_pressed_special(key, x, y):
    if key == GLUT_KEY_UP:
        glTranslatef(0.0, 10.0, 0.0)
    elif key == GLUT_KEY_DOWN:
        glTranslatef(0.0, -10.0, 0.0)
    elif key == GLUT_KEY_RIGHT:
        glTranslatef(10.0, 0.0, 0.0)
    elif key == GLUT_KEY_LEFT:
        glTranslatef(-10.0, 0.0, 0.0)
    elif key == GLUT_KEY_PAGE_UP:
        glTranslatef(0.0, 0.0, 10.0)
    elif key == GLUT_KEY_PAGE_DOWN:
        glTranslatef(0.0, 0.0, -10.0)

def key_pressed(key, x, y):
    global angle_x, angle_y, angle_z
    
    if key == b'x':
        angle_x = (angle_x+10)%360
    elif key == b'X':
        angle_x = (angle_x-10)%360
    elif key == b'y':
        angle_y = (angle_y+10)%360
    elif key == b'Y':
        angle_y = (angle_y-10)%360
    elif key == b'z':
        angle_z = (angle_z+10)%360
    elif key == b'Z':
        angle_z = (angle_z-10)%360
    elif key == b's':
        glScalef(0.5, 0.5, 0.5)
    elif key == b'S':
        glScalef(2.0, 2.0, 2.0)
```

## Funções auxiliares

```python
def window_resize(w, h):
    # Função chamada quando a janela é reescalonada
    if(h == 0):
        h = 1
    
    glViewport(0, 0, w, h)

    fAspect = w/h
    angle = 45

    glMatrixMode(GL_PROJECTION)

    glLoadIdentity()

    gluPerspective(angle,fAspect,0.1,10000)

    glMatrixMode(GL_MODELVIEW)
    
    glLoadIdentity()

    gluLookAt(0,50,200, 0,0,0, 0,1,0)


def timer(value):
    # Função usada para chamar display() a cada 33ms
    glutPostRedisplay()
    glutTimerFunc(33, timer, 0)
```

## Função init
Na função init, são definidos os parâmetros para a iluminação da cena, o que inclui as propriedades dos materias e da luz.

```python
def init():   
    luzAmbiente = [0.2,0.2,0.2,1.0]
    luzDifusa = [0.7,0.7,0.7,1.0]
    luzEspecular = [1.0, 1.0, 1.0, 1.0]
    posicaoLuz = [0.0, 50.0, 50.0, 1.0]

    especularidade = [1.0,1.0,1.0,1.0]
    especMaterial = 60

    glClearColor(0.0, 0.0, 0.0, 1.0)

    glShadeModel(GL_SMOOTH)

    glMaterialfv(GL_FRONT,GL_SPECULAR, especularidade)
    glMateriali(GL_FRONT,GL_SHININESS,especMaterial)

    glLightModelfv(GL_LIGHT_MODEL_AMBIENT, luzAmbiente)

    glLightfv(GL_LIGHT0, GL_AMBIENT, luzAmbiente)
    glLightfv(GL_LIGHT0, GL_DIFFUSE, luzDifusa)
    glLightfv(GL_LIGHT0, GL_SPECULAR, luzEspecular)
    glLightfv(GL_LIGHT0, GL_POSITION, posicaoLuz)

    glEnable(GL_COLOR_MATERIAL)
    glEnable(GL_LIGHTING)
    glEnable(GL_LIGHT0)
    glEnable(GL_DEPTH_TEST)
```

## Registro das funções do glut e main loop

```python
glutInit(sys.argv)
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowSize(800, 600)
glutInitWindowPosition(0, 0)
glutCreateWindow("Pentagon")
glutDisplayFunc(display)
glutTimerFunc(33, timer, 0)
glutReshapeFunc(window_resize)
glutMouseFunc(mouse_click)
glutKeyboardFunc(key_pressed)
glutSpecialFunc(key_pressed_special)
init()

glutMainLoop()
```
